## 0. Update key value funtion

```python
def update_variable(variable_name,variable, player_dict):
    player_dict[str(variable_name)]=variable
```

## Initalize Dictonary

```python
p_name="A Mishra"
player_dict={"name":p_name}
player_dict["type"]="bowler"
player_dict
```

```python
def get_pdf_b(p_name):
    pdf=df[df["bowler"]==p_name]
    return pdf
pdf=get_pdf_b(p_name)
```

## Get Origin and espncricinfo.com id

```python
def origin_id(p_name, pdm, player_dict):
    origin=pdm.loc[p_name]["Origin"]
    profile_id=pdm.loc[p_name]["Profile Id"]
    update_variable("origin",origin, player_dict)
    update_variable("profile_id",profile_id, player_dict)
    return player_dict

player_dict=origin_id(p_name, pdm, player_dict)
```

## 1. Strike_rate as bowler


```python
pdf
```

```python
def strike_rate_b(pdf, player_dict):
    total_balls=pdf.count()["ball"]
    
    total_wicket=pdf.count()["player_dismissed"]
    strike_rate_no=total_balls/total_wicket
    
    update_variable("strike_rate",strike_rate_no, player_dict)

strike_rate_b(pdf, player_dict)
```

## 2. Economy

```python
pdf.count()["ball"]/6
```

```python
def economy(pdf, player_dict):
    total_runs=pdf.sum()["total_runs"]
    total_overs=pdf.count()["ball"]/6
    economy=total_runs/total_overs
    economy=round(economy,2)
    update_variable("economy",economy, player_dict)
    
economy(pdf, player_dict)
```

## 3. No of 4 or 5 wickets

```python
def total_3_4_b(pdf, player_dict):
    x=pdf.groupby("match_id").count()
    total_3_b=len(x[x["player_dismissed"]==3])
    total_4_b=len(x[x["player_dismissed"]==4])
    update_variable("3w",total_3_b, player_dict)
    update_variable("4w",total_4_b, player_dict)
total_3_4_b(pdf, player_dict)
```

## 4. No. Of Matches Played

```python
pdf.groupby("match_id").count()
```

```python
def no_of_matches(pdf, player_dict):
    no_of_matches=len(pdf.groupby("match_id").count())
    update_variable("no_of_matches",no_of_matches, player_dict)

no_of_matches(pdf, player_dict)
```

# 5. Winning Percentage

```python
mf.head(1)
```

```python
list(pdf.groupby("match_id").count().reset_index()["match_id"])
```

```python
def update_match_status(pdf, p_name):
    
    print("Match Status Started for:",p_name)
    match_ids=list(pdf.groupby("match_id").count().reset_index()["match_id"])
    
    
    for m_id in match_ids:
        
        player_team  =  pdf[pdf["match_id"]==m_id].iloc[0]["bowling_team"]
        winning_team =  mf.loc[m_id]["winner"]

        if winning_team==player_team:
            status="Winner"
        elif pd.isnull(winning_team):
            status="No Result"
        else:
            status="Looser"
            
            
        pdf.loc[pdf["match_id"]==m_id, "match_status"]=status

update_match_status(pdf, p_name)
```

```python
def winning_percentage_b(pdf, player_dict):
    first_element_of_group=pdf.groupby(["match_id","match_status"]).first().reset_index()
    frequency=first_element_of_group["match_status"].value_counts()
    try:
        won=frequency["Winner"]
    except:
        won=0
    try:
        lost=frequency["Looser"]
    except:
        lost=0
    winning_percentage=100*won/(won+lost)
    winning_percentage=round(winning_percentage,2)
    update_variable("winning_percentage_b",winning_percentage, player_dict)

winning_percentage_b(pdf, player_dict)
```

```python
def most_wickets(pdf, player_dict):
    most_wickets=max(pdf.groupby("match_id").count()["player_dismissed"])
    update_variable("most_wickets",most_wickets, player_dict)
most_wickets(pdf, player_dict)
```

```python
def update_bins_b(pdf,player_dict):
    value=pdf.groupby("match_id").count()["ball"].values
    value=value/6
    value=np.round(value,decimals=2)

    bins = [1,2,4,6]
    hist=np.histogram(value,bins)
    
    for h in range(0,4):
        key_name=str(hist[1][h])+"os"
        
        try: 
            value=hist[0][h]
        except:
            value=0
        
        player_dict[key_name]=value
```

## Creating aggregator function

```python
def complete_agrregator(p_name, df, mf, pdm):
    
    # Intialize Dict
    player_dict={"name":p_name}

    #Profile Type
    player_dict["type"]="bowler"
    
    #Get player Data
    pdf=get_pdf_b(p_name)
    

    # Get origin and Espencrickinfo.com id
    player_dict=origin_id(p_name, pdm, player_dict)
    
    #     1. Strike Rate
    strike_rate_b(pdf, player_dict)
    
    #2. Economy
    economy(pdf, player_dict)
    
    #3. No Of 4, 5 wickets 
    total_3_4_b(pdf, player_dict)
    
    #4. No of Matches PLayed
    no_of_matches(pdf, player_dict)
    
    
    #5. Win /Loose
    update_match_status(pdf, p_name)
    winning_percentage_b(pdf, player_dict)
    
    #6. Strenght player
    dismissed_cols=df[df["dismissal_kind"].notnull()]["dismissal_kind"].unique()
    player_dict=dismissal_type_b(pdf,dismissed_cols,player_dict)
    
    #7. Most wickets
    most_wickets(pdf, player_dict)
    
    #8. Will leave after (Over bins)
    update_bins_b(pdf,player_dict)


    return player_dict
```

# Multi Threading  for faster results

```python
import threading
import math

#### Importing Csv File
df=pd.read_csv("./input/deliveries.csv")
mf=pd.read_csv("./input/matches.csv", index_col="id")
pdm=pd.read_csv("./output/till_orgin_id.csv", index_col="Name")


## Creating Main Dictonary
player_dicts=[]

## DIsrtibutiing workload 
player_list=list(set(df["bowler"]))
total_players=len(player_list)


## Setting Processor Power 
thread_count=8
delta=total_players/thread_count


## Making thread list
thread_list=[]
for i in range(thread_count):
    start=math.floor(i*delta)
    end=math.floor( (i+1)*delta)
    
    ## We are dealing with python list which starts form 0 and not 1
    list_end=total_players-1
    
    if end > list_end:
        end=list_end
    
    
    t=threading.Thread(target=save_all, args=(player_list[start:end],))
    thread_list.append(t)
    
    
    print("Start:", start, "End:", end)
    print("Start:", player_list[start], "End:", player_list[end])
```

```python
# Starting thread now
```

```python
## Starting Thread
import time

start_time=time.time()

for t in thread_list:
    t.start()


## Rejoining Thread
for t in thread_list:
    t.join()

end_time=time.time()
```
