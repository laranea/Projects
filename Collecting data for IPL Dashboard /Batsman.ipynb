## 0. Update key value funtion

```python
def update_variable(variable_name,variable, player_dict):
    player_dict[str(variable_name)]=variable
    
```

## Initalize Dictonary

```python
p_name="CH Gayle"
player_dict={"name":p_name}
player_dict["type"]="batsman"
player_dict
```

```python
def get_pdf(p_name):
    pdf=df[df["batsman"]==p_name]
    return pdf

def get_pdf_s(p_name): 
    pdf_s=df[(df["batsman"]==p_name) | (df["non_striker"]==p_name) ]
    return pdf_s
```

```python
pdf=get_pdf(p_name)
pdf_s=get_pdf_s(p_name)
```

## Get Origin and espncricinfo.com id

```python
def origin_id(p_name, pdm, player_dict):
    origin=pdm.loc[p_name]["Origin"]
    profile_id=pdm.loc[p_name]["Profile Id"]
    update_variable("origin",origin, player_dict)
    update_variable("profile_id",profile_id, player_dict)
    return player_dict

player_dict=origin_id(p_name, pdm, player_dict)    
```

```python
player_dict
```

## 1. Strike Rate

```python
def strike_rate(pdf, player_dict):
    total_runs=sum(pdf["total_runs"])
    total_balls=pdf.shape[0]
    strike_rate=100*(total_runs/total_balls)
#     print(strike_rate)
    update_variable("strike_rate",strike_rate, player_dict)

### Saving Data
strike_rate(pdf, player_dict)
player_dict
```

## 2. Batting Average

```python
def batting_average(pdf, pdf_s, p_name, player_dict):
    
    total_runs=sum(pdf["total_runs"])
    times_dismised=pdf_s[pdf_s["player_dismissed"]==p_name].count()["player_dismissed"]
    
    batting_average=total_runs/times_dismised
    
#     print(batting_average)
    update_variable("batting_average",batting_average, player_dict)

### Saving Data
batting_average(pdf, pdf_s,p_name,player_dict)
player_dict
```

```python
total_runs=sum(pdf["total_runs"])
total_runs
pdf_s
```

## 3. No. of 4 and 6


```python
def total_4_6(pdf, player_dict):
    total_4=pdf[pdf["total_runs"]==4].count()["total_runs"]
#     print(total_4)
    total_6=pdf[pdf["total_runs"]==6].count()["total_runs"]
#     print(total_6)
    update_variable("4s",total_4, player_dict)
    update_variable("6s",total_6, player_dict)
        
### Saving Data
total_4_6(pdf, player_dict)
```

## 4. No. of Matches Played

```python
def no_of_matches(pdf_s, player_dict):
    no_of_matches=len(set(pdf_s["match_id"]))
#     print(no_of_matches)
    update_variable("no_of_matches",no_of_matches, player_dict)

no_of_matches(pdf_s, player_dict)
```

## 5. Wining percentage

```python
mf.head(1)
```

```python
def update_match_status(pdf_s, p_name):
    
    print("Match Status Started for:",p_name)
    match_ids=list(set(pdf_s["match_id"]))
    
    for m_id in match_ids:
        
        player_team  =  pdf_s[pdf_s["match_id"]==m_id].iloc[0]["batting_team"]
        winning_team =  mf.loc[m_id]["winner"]

        if winning_team==player_team:
            status="Winner"
        elif pd.isnull(winning_team):
            status="No Result"
        else:
            status="Looser"
            
            
        pdf_s.loc[pdf_s["match_id"]==m_id, "match_status"]=status
    
```

```python
update_match_status(pdf_s, p_name)
```

```python
pdf_s.head(1)
```

```python
def winning_percentage(pdf_s, player_dict):
    first_element_of_group=pdf_s.groupby(["match_id","match_status"]).first().reset_index()
    frequency=first_element_of_group["match_status"].value_counts()
    try: 
        won=frequency["Winner"]
    except:
        won=0
    try:    
        lost=frequency["Looser"]
    except:
        lost=0
    winning_percentage=100*won/(won+lost)
    winning_percentage=round(winning_percentage,2)
    update_variable("winning_percentage",winning_percentage, player_dict)
```

```python
winning_percentage(pdf_s, player_dict)
```

```python
dismissed_cols=df[df["dismissal_kind"].notnull()]["dismissal_kind"].unique()
### Saving Data

def dismissal_type(pdf,dismissed_cols,player_dict):
    for col in dismissed_cols:
        try:
            player_dict[col]=pdf["dismissal_kind"].value_counts()[col]
        except:
            player_dict[col]=0
            pass
```

## 7. Most Runs

```python
def most_runs(pdf, player_dict):
    most_runs=max(pdf.groupby("match_id").sum()["total_runs"])

#     print(most_runs)
    ### Saving Data
    update_variable("most_runs",most_runs, player_dict)
most_runs(pdf, player_dict)
```

## 8.  Out After (Run Bins)

```python
def update_bins(pdf,player_dict):
    value=pdf.groupby("match_id").sum()["total_runs"].values
    bins = [0,25,50,75,200]
    hist=np.histogram(value,bins)
    
    print(hist)
    for h in range(0,4):
        key_name=str(hist[1][h])+"-"+str(str(hist[1][h+1])+"s")
        
        try: 
            value=hist[0][h]
        except:
            value=0
        
        player_dict[key_name]=value
    
```

```python
update_bins(pdf,player_dict)
player_dict
```

## Creating aggregator function

```python
def complete_agrregator(p_name, df, mf, pdm):
    
    # Intialize Dict
    player_dict={"name":p_name}

    #Profile Type
    player_dict["type"]="batsman"
    
    #Get player Data
    pdf=get_pdf(p_name)
    pdf_s=get_pdf_s(p_name)
    
    # Get origin and Espencrickinfo.com id
    player_dict=origin_id(p_name, pdm, player_dict)
    
    
    #1. Strike Rate
    strike_rate(pdf, player_dict)
    
    #2. Batting Average
    batting_average(pdf,pdf_s, p_name, player_dict)
    
    #3. No Of 4, 6
    total_4_6(pdf, player_dict)
    
    #4. No of Matches PLayed
    no_of_matches(pdf_s, player_dict)
    
    
    #5. Win /Loose
    update_match_status(pdf_s, p_name)
    winning_percentage(pdf_s, player_dict)
    
    #6. How to defeat him
    dismissed_cols=df[df["dismissal_kind"].notnull()]["dismissal_kind"].unique()
    dismissal_type(pdf,dismissed_cols,player_dict)
    
    #7. Most run
    most_runs(pdf, player_dict)
    
    #8. Out After (Run Bins)
    update_bins(pdf,player_dict)

    return player_dict
```

```python
def save_all(player_list):

    print("Started thread from:",player_list[0], "to", player_list[-1])
        
    
#     player_list=bat_set=set(df["batsman"])

    for player in player_list:
        p_name=str(player)
        player_dict=complete_agrregator(p_name,df, mf, pdm)
        player_dicts.append(player_dict)

    return player_dicts
```

# Multi Threading  for faster results

```python
import threading
import math

#### Importing Csv File
df=pd.read_csv("./input/deliveries.csv")
mf=pd.read_csv("./input/matches.csv", index_col="id")
pdm=pd.read_csv("./output/till_orgin_id.csv", index_col="Name")


## Creating Main Dictonary
player_dicts=[]

## DIsrtibutiing workload 
player_list=list(set(df["batsman"]))
total_players=len(player_list)


## Setting Processor Power 
thread_count=8
delta=total_players/thread_count


## Making thread list
thread_list=[]
for i in range(thread_count):
    start=math.floor(i*delta)
    end=math.floor( (i+1)*delta)
    
    ## We are dealing with python list which starts form 0 and not 1
    list_end=total_players-1
    
    if end > list_end:
        end=list_end
    
    
    t=threading.Thread(target=save_all, args=(player_list[start:end],))
    thread_list.append(t)
    
    
    print("Start:", start, "End:", end)
    print("Start:", player_list[start], "End:", player_list[end])
```

## Starting thread now

```python
## Starting Thread
import time

start_time=time.time()

for t in thread_list:
    t.start()


## Rejoining Thread
for t in thread_list:
    t.join()

end_time=time.time()
```
