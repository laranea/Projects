```r
raw_df=read_excel(
    path=system.file(package="ScPoEconometrics", "datasets", "demo_gind.xls"),
    sheet="Data",
    range="A9:K68"
)
raw_df
```

```r
?gather
```

```r
library(tidyr)
df=gather(raw_df , paste(2008:2017), key="year", value="counts")
df
```

```r
df$counts = as.integer(df$counts)
df
```

```r
library(dplyr)

df %>%
    
    filter(Country %in% c("France", "Italy", "Ukraine")) %>%
    
    mutate(millions=counts/1e6) %>%

    ggplot(
        mapping=aes(
                    x=year,
                    y=millions,
                    color=Country,
                    group=Country
            
                    )
    
        )+geom_line(size=1)
```

```r
top5= df %>%
    group_by(Country)%>%
#     mutate(mean_count=mean(counts)) %>%
    summarise(mean_count=mean(counts)) %>%
    arrange(desc(mean_count))

top5
```
