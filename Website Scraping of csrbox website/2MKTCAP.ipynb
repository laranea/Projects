```python
from difflib import SequenceMatcher

def dimilar_diff(a, b):
    a1=a.replace(" ", "")
    b1=b.replace(" ", "")
    r=SequenceMatcher(None, a1, b1).ratio()
    if r > .80:
        print("%s : %s --- %s"% (r, a1, b1))
        return True
    else:
        return False
```

```python
import Levenshtein

def similar(a, b):
    a1=a.replace(" ", "")
    b1=b.replace(" ", "")
    r=Levenshtein.ratio(a1, b1)
    if r > .75:
        print("%s : %s --- %s"% (r, a1, b1))
        return r
    else:
        return -2
```
