```python
def retrive_value(company_url):
    url=company_url
    html=urlopen(url)
    soup=BeautifulSoup(html,'lxml')
    
    title=soup.title
    clean_title=title.get_text()
    print(clean_title)
    
    divs=soup.findAll("div", {"class":"dfy"})
    divs =str(divs)
    
    clean_div=BeautifulSoup(divs, "lxml").get_text()
    
    ## Remove first 3 items
    div_list=clean_div.split(",")[3:]
    
    # Break list in case of repeating daata from div
    div_list=div_list[:break_list(div_list)]
    
    
    ## Remove all BS spaces as they keep fing us
    div_list = [x.replace(" ", "") for x in div_list]

    
    print(div_list)
    
    """
    Specifying cols and Values here
    """
    # Initialize list
    df_cols=["Company"]
    df_values=[clean_title]


    for item in div_list:
        if "-" in item:
            df_cols.append(str(item)+" Actual")
            df_cols.append(str(item)+" Prescribed")
        else:
            df_values.append(item)
        print(df_cols)
        print(df_values)


    company_values_s=pd.Series(df_values, index=df_cols)
    
    return company_values_s
```

```python
for u in company_list_u:
    company_value_s=retrive_value(u)
    df=df.append(company_value_s, ignore_index=True)
```
