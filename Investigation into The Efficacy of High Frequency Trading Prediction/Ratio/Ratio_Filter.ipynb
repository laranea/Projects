```python
ratio_status=[]
common_ratio=main_ratios

for t in common_ticks:
    
    ratio_length=len(common_ratio)
    ratio_status.append(ratio_length)
    
    
#   Some Info for comman man
    print("Check for : {0}".format(t) )
    print("U_R Length: {0}".format(ratio_length))
    
#   Find query and convert to list    
    data=ratio.loc[ratio["tick"]==t, "u_ratio"]
    data=data.tolist()
    common_ratio= set(common_ratio) & set(data)
    common_ratio=list(common_ratio)    
    
    
```
